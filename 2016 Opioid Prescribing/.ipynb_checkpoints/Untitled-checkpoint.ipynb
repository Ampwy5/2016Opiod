```python
import pandas as pd
import numpy as np
```

```python
county_data = "county_df.csv"
```

```python
opiates = pd.read_csv(county_data)
```

```python
opiates
```

```python
labor_data = "merged_labor_data.csv"
```

```python
labor = pd.read_csv(labor_data)
```

```python
labor
```

```python
labor = labor.rename(columns={'County Name': 'County Name'})
opiates = opiates.rename(columns={'County Name': 'County Name'})
```

```python
labor['County Name'] = labor['County Name'].str.strip()
labor['State Abbreviation'] = labor['State Abbreviation'].str.strip()
opiates['County Name'] = opiates['County Name'].str.strip()
opiates['State Abbreviation'] = opiates['State Abbreviation'].str.strip() 
```

```python
labor = labor.dropna()
labor
```

```python
opiates
```

```python
merge = pd.merge(labor, opiates, on=["County Name", "State Abbreviation"])
```
